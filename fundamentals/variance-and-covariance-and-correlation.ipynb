# Variance, covariance and correlation

## Variance

- The variance is a measure of variability of a "random variable".
    - In statistics, a random variable is a mathematical concept that represents a quantity whose value is determined by the outcome of a random trial or experiment.
    - Practically, it is a series of "numeric results", like heads/tails from tossing a coin, daily closing price of NVDA, temperature reading per second, etc.


- The definition of variance is:

    $$
    Var(X)=\frac{\sum_{i=1}^n(x_i-\mu)^2}{n}
    $$
    Where $\mu$ is the mean of $x_0$, $x_1$, $x_2$, ... , $x_n$.
    - Variance is also denoted by $\sigma^2_X$ where $X$ is a random variable.

- Expected value of a random variable
  - The expected value of a random variable is the weighted average of all possible values of the variable.
  - The expected value of a random variable $X$ is often denoted by $E(X)$, $E[X]$, or $EX$, with $E$ also often stylized as ${\displaystyle \mathbb {E}}$.
  - For example, Let ${\displaystyle X}$ represent the outcome of a roll of a fair six-sided die. The possible values for $X$ are 1, 2, 3, 4, 5, and 6, all of which are equally likely with a probability of $\frac{1}{6}$. The expectation of ${\displaystyle X}$ is 
  $$
  E[X] = 1 * \frac{1}{6} + 2 * \frac{1}{6} +  3 * \frac{1}{6} + 4 * \frac{1}{6} + 5 * \frac{1}{6} + 6 * \frac{1}{6} = 3.5
  $$

- Alternatively, the variance of a random variable $X$ can also be defined as the expected value of the squared deviation from the mean of ${\displaystyle X}$:

  $$
  {\displaystyle \operatorname {Var} (X)=\operatorname {E} \left[(X-\mu )^{2}\right]}
  $$
  where $\mu = E[X]$

```python
import numpy as np

X = [10, 15, 20, 25, 30]
μ = np.mean(X)
σ_squared_X = np.sum((np.array(X) - μ) ** 2) / len(X)

print(f"X: {X}")
print(f"μ: {μ}")
print(f"σ_squared_X: {σ_squared_X}")
```

## Covariance

- Covariance is the variance between two random variables.

  $$
  Cov(X, Y)=\frac{\sum_{i=1}^n(x_i-\bar{x})(y_i-\bar{y})}{n}
  $$


```python
import numpy as np

# Sample data
X = [10, 15, 20, 25, 30]
Y = [5, 10, 15, 20, 25]

# Calculate the means
μ_x = np.mean(X)
μ_y = np.mean(Y)

# Calculate the covariance
covariance = np.sum((np.array(X) - μ_x) * (np.array(Y) - μ_y)) / len(X)

print("Sample data (x):", X)
print("Sample data (y):", Y)
print("Mean of x:", μ_x)
print("Mean of y:", μ_y)
print("Covariance:", covariance)
```

### Covariance matrix

- If we have more than two random variables (e.g., $X_1$, $X_2$, $X_3$), instead of a covariance scalar, we will have a covariance matrix.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate some sample data
np.random.seed(43)
data = np.random.normal(0, 1, size=(100, 5))

df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])
print(df)

# Calculate the covariance matrix
cov_matrix = df.cov(ddof=0)

# Print the covariance matrix
print("Covariance Matrix:")
print(cov_matrix)

# Visualize the covariance matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, cmap='YlOrRd')
plt.title('Covariance Matrix Heatmap')
plt.xlabel('Variables')
plt.ylabel('Variables')
plt.show()
```

## Correlation

- In statistics, correlation is any statistical relationship, between two random variables. 

    $$
    corr(X, Y) = \frac{cov(X, Y)}{\sigma_X\sigma_Y}
    $$

    where $\sigma_X$ is the standard deviation of X (in contrast to $\sigma^2_X$, meaning the variance of $X$)

    - $corr(X, Y)$ is also called Pearson correlation if both $\sigma_X$ and $\sigma_Y$ are finite and positive.

```python
import numpy as np

X = [10, 15, 20, 25, 30]
μ_X = np.mean(X)
σ_squared_X = np.sum((np.array(X) - μ_X) ** 2) / len(X)

Y = [3, 1, 4, 1 ,5]
μ_Y = np.mean(Y)
σ_squared_Y = np.sum((np.array(Y) - μ_Y) ** 2) / len(Y)

cov_XY = np.sum((np.array(X) - μ_X) * (np.array(Y) - μ_Y)) / len(X)

corr = cov_XY / (σ_squared_X ** 0.5 * σ_squared_Y ** 0.5)
print(corr)
```
